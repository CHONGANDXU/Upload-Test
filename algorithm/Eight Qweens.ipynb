**算法 1：** 为解决 8 皇后问题，初步想法为，以随机数挑选皇后棋子的位置，一层一层判断是否符合要求


```python
import random

print("A possile solution is:")
i = range(0, 8)
arr_change = list(i)
arr_ran = list()
for m in range(0, 8):
    if m == 0:
        n = random.randint(0, 7 - m)
        arr_ran.append(arr_change[n])
        arr_change.remove(arr_change[n])

    if m >= 1 and m <= 6:
        Flag = False
        while not Flag:
            n = random.randint(0, 7 - m)
            arr_ran.append(arr_change[n])
            for j in range(0, m):
                if abs(arr_ran[m] - arr_ran[j]) == (m - j):
                    Flag = False
                    arr_ran.remove(arr_change[n])
                    break
                else:
                    Flag = True
        arr_change.remove(arr_change[n])

    if m == 7:
        arr_ran.append(arr_change[0])
        arr_change.remove(arr_change[0])
    # print(arr_ran)
    # print(arr_change)
    ran = arr_ran[m]
    for i in range(0, 8):
        if i == ran:
            print("|Q", end="")
        else:
            print("|", end=" ")
    print("|")

# 以上算法会陷入死循环，倘若最后两个随机数放入时，都不符合条件，则陷入死循环，没有跳出，报错
```

**算法 2：**

- 随机生成一个数列，直接判断其数列符合要求与否
- 倘若数列不符合要求，则推翻此数列，再重新随机生成


```python
import random

print("A possile solution is:")
Flag = False
while not Flag:
    ran = range(0, 8)
    arr_change = list(ran)
    arr_ran = list()
    for m in range(0, 8):
        n = random.randint(0, 7 - m)
        arr_ran.append(arr_change[n])
        arr_change.remove(arr_change[n])
    for i in range(1, 8):
        for j in range(0, i):
            if abs(arr_ran[i] - arr_ran[j]) == (i - j):
                Flag = False
                break
            else:
                Flag = True
        if not Flag:
            break
print("最终满足要求的位置为", arr_ran)
for i in range(0, 8):

    for j in range(0, 8):
        if j == arr_ran[i]:
            print("|Q", end="")
        else:
            print("|", end=" ")
    print("|")
```

**算法3**:
    标准解决方法

```python
def conflict(state, nextColumn):
    nextRow = rows = len(state)
    for row in range(rows):
        column = state[row]
        if abs(column - nextColumn) in (0, nextRow - row):
            """
            如果差值等于0,两个皇后在同一列,则代表冲突,返回True;
            如果列的差值等与行的差,两个皇后在对角线上,则代表冲突,返回True;
            """
            return True
    return False


def queens(num, state=()):
    """
    采用生成器的方式来产生每一个皇后的位置,并用递归来实现下一个皇后的位置
    num: 皇后的数量
    state: 标记已经排好的每个皇后的位置
    """
    for pos in range(num):  # 八皇后的数量N=0, 1, 2, 3, 4, 5, 6 , 7 你要在哪一列放置皇后
        # 如果不冲突，则递归构造棋盘。
        if not conflict(state, pos):  # 回溯法的体现
            # 如果棋盘状态state已经等于num-1，即到达倒数第二行，而这时最后一行皇后又没冲突，直接yield，打出其位置(pos, )
            if len(state) == num - 1:  # state=()
                yield (pos,)
            else:  # (0, )
                for result in queens(num, state + (pos,)):
                    yield (pos,) + result


def prettyprint(solution):

    def line(pos, length=len(solution)):
        return "| " * (pos) + "|Q" + "| " * (length - pos - 1) + "|"

    for pos in solution:
        print(line(pos))


if __name__ == "__main__":
    solutions = queens(8)
    for index, solution in enumerate(solutions):
        print("第%d种解决方案:" % (index + 1), solution)
        prettyprint(solution)
        print("\n" + "*" * 50 + "\n")
```
